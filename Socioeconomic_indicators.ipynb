```python
import pandas as pd

file_path = "/content/Census_Data_-_Selected_socioeconomic_indicators_in_Chicago,_2008_â€“_2012_20250920.csv"

try:
    df = pd.read_csv(file_path)
    display(df.head())
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    print("Please make sure the file is uploaded or the correct path is provided.")
```

```python
#How many rows are in the dataset?
num_rows = df.shape[0]
print(f"The dataset has {num_rows} rows.")
```

```python
hardship_above_50 = df[df['HARDSHIP INDEX'] > 50.0]
num_community_areas = hardship_above_50.shape[0]
print(f"Number of community areas with hardship index greater than 50.0: {num_community_areas}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['PER CAPITA INCOME '], df['HARDSHIP INDEX'])
plt.xlabel('Per Capita Income')
plt.ylabel('Hardship Index')
plt.title('Per Capita Income vs. Hardship Index in Chicago Community Areas')
plt.grid(True)
plt.show()
```
